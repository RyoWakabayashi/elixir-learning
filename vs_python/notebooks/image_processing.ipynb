```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import requests
```

```python
def display_image(target_img):
    plt.imshow(cv2.cvtColor(target_img, cv2.COLOR_BGR2RGB))
    plt.show()
```

## 画像のダウンロード、表示

```python
response = requests.get("https://www.elixirconf.eu/assets/images/ryo-wakabayashi.png")

print(response)

img_array = np.asarray(bytearray(response.content), dtype=np.uint8)

print(img_array)

img = cv2.imdecode(img_array, cv2.IMREAD_COLOR)
```

```python
display_image(img)
```

## リサイズ

```python
resized_img = cv2.resize(img, [600, 400])

display_image(resized_img)
```

## グレースケール

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

display_image(gray_img)
```

## ぼかし

```python
blur_img = cv2.blur(img, [15, 15])

display_image(blur_img)
```

## 図形描画

```python
drawed_img = cv2.rectangle(img,
  # 左上{x, y}
  [150, 120],
  # 右下{x, y}
  [225, 320],
  # 色{R, G, B}
  [0, 0, 255],
  # 線の太さ
  12,
  # 線の引き方（角がギザギザになる）
  cv2.LINE_4
)

display_image(drawed_img)
```

## 文字描画

```python
drawed_img = cv2.putText(img,
  # 文字列
  "Ryo",
  # 左下{x, y}
  [250, 200],
  # フォント種類
  cv2.FONT_HERSHEY_SIMPLEX,
  # フォントサイズ
  2.5,
  # 文字色
  [0, 0, 255],
  # 文字太さ
  5
)

display_image(drawed_img)
```

## 一連処理

```python
processed_img = cv2.resize(img, [600, 400])

processed_img = cv2.cvtColor(processed_img, cv2.COLOR_BGR2GRAY)

processed_img = cv2.blur(processed_img, [15, 15])

processed_img = cv2.rectangle(processed_img,
  # 左上{x, y}
  [150, 120],
  # 右下{x, y}
  [225, 320],
  # 色{R, G, B}
  [0, 0, 255],
  # 線の太さ
  12,
  # 線の引き方（角がギザギザになる）
  cv2.LINE_4
)

processed_img = cv2.putText(processed_img,
  # 文字列
  "Ryo",
  # 左下{x, y}
  [250, 200],
  # フォント種類
  cv2.FONT_HERSHEY_SIMPLEX,
  # フォントサイズ
  2.5,
  # 文字色
  [0, 0, 255],
  # 文字太さ
  5
)

display_image(processed_img)
```
